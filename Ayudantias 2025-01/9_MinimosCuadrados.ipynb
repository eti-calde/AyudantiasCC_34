# **Mínimos Cuadrados**

## **Contexto**

Suponga que hacemos un experimento en el que le pedimos a cada estudiante al dar un certamen que anote la cantidad de horas que estudio para el mismo. Lo que nos genera una lista de puntos $(x_i, y_i)$ donde:

- $x$ es la nota del estudiante.
- $y$ es la cantidad de horas estudiadas.

Tenemos sospechas de que debería existir un relación lineal entre horas y notas. pero además de que nuestra data esta manchada, ya sea por estudiantes que mintieron sobre la cantidad de horas o que no midieron correctamente la cantidad de horas.

Nuestro objetivo NO es encontrar una curva que pase exactamente por cada punto, sino una recta que se aproxime lo mejor posible a todos ellos. Eso es lo que hace mínimos cuadrados.

#### **PARAMETROS GENERALES**

```python
np.random.seed(42)
m = 60  # Número de estudiantes
```

#### **Datos Sinteticos**

```python
x = np.random.uniform(0, 100, m)  # Horas de estudio entre 0 y 100
y_real = 2 + 0.5 * x  # Relación real (sin error)
y = y_real + 5*np.random.normal(0, 1, m)  # Notas con error

# Graficar datos
plt.scatter(x, y, label='Datos simulados')
plt.plot(x, y_real, color='red', linestyle='--', label='Modelo real (y=2+0.5x)')
plt.xlabel('Horas de estudio')
plt.ylabel('Nota')
plt.legend()
plt.show()
```
